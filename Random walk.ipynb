```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
m = 100
size = 30
g = -size
p = np.zeros([m]) # 統計N顆粒子的位置分布
q = np.zeros([n]) # 每次Random Walk N次後，最後的位置
for i in range(n): # 計算每顆粒子最後的位置，並存入q中
    for j in range(n):
        x = np.random.random()        
        q[i] = q[i] + x - 0.5
for k in range(m): #劃分區間，並統計每個區間內的粒子個數
    for l in range(n):
        if q[l] >= g and q[l] < g + ( 2 * size / n ): #統計-size到-size+2*size/n區間內的粒子個數
            p[k] = p[k] + 1
    g = g + ( 2 * size / m )
plt.figure(figsize=(20, 10))
plt.subplot(1,2,1)
plt.plot(q,'.b')
plt.title("1D Random Walk")
plt.xlabel('Orders')
plt.ylabel('Position')
plt.subplot(1,2,2)
plt.hist(q, bins= m, linewidth=1)
plt.title("1D Random Walk")
plt.xlabel('Position')
plt.ylabel('Times')
plt.show()
```

```python
n = 1000
size = 60
x = np.zeros([n]) # 每次Random Walk N次後，最後的位置
y = np.zeros([n]) # 每次Random Walk N次後，最後的位置
for i in range(n):
    for j in range(n):
        a = np.random.random()
        b = np.random.random()
        if a > 0.5 and b > 0.5:
            x[i] = x[i] + 0.5
            y[i] = y[i] + 0.5
        elif a > 0.5 and b <= 0.5:
            x[i] = x[i] + 0.5
            y[i] = y[i] - 0.5
        elif a <= 0.5 and b > 0.5:
            x[i] = x[i] - 0.5
            y[i] = y[i] + 0.5
        else :
            x[i] = x[i] - 0.5
            y[i] = y[i] - 0.5
plt.figure(figsize=(10, 10))
plt.plot(x,y,'.b')
plt.title("2D Random Walk_Method 1")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.xlim(-size, size)
plt.ylim(-size, size)
plt.xticks(np.linspace(-size, size, 13, endpoint=True))
plt.yticks(np.linspace(-size, size, 13, endpoint=True))
plt.show()
```

```python
n = 1000
size = 60
x = np.zeros([n]) # 每次Random Walk N次後，x最後的位置
y = np.zeros([n]) # 每次Random Walk N次後，y最後的位置
for i in range(n):
    for j in range(n):
        a = np.random.random() - 0.5
        b = np.random.random() - 0.5
        x[i] = x[i] + a
        y[i] = y[i] + b
plt.figure(figsize=(10, 10))
plt.plot(x,y,'.b')
plt.title("2D Random Walk_Method 2")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.xlim(-size, size)
plt.ylim(-size, size)
plt.xticks(np.linspace(-size, size, 13, endpoint=True))
plt.yticks(np.linspace(-size, size, 13, endpoint=True))
plt.show()
```
